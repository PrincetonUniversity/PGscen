```python
import os
import pandas as pd
from pathlib import Path
```

# Wind

```python
wind_mapping = pd.read_csv('~/Downloads/Wind_Mapping_NY_NREL2030.csv')
wind_meta_df = pd.read_csv('~/search/PERFORM/Data/ForArnab/MetaData/wind_meta.csv')
```

```python
datadir = '~/Research/PERFORM/Data/ForArnab/'
savedir = '~/Research/PERFORM/Data/ForArnab/NYISO/'

for date in pd.date_range(start='2019-08-02', periods=7, freq='D'):
#     date.strftime('%Y%m%d')
    
    for _, row in wind_mapping.iterrows():
        ny_name = row['name_NY']
        
        nrel_name = row['name_NREL']
        if nrel_name == 'Dutch Hill Cohocton':
            nrel_name = 'Dutch Hill/Cohocton'
        
        ny_cap = row['capacity_NY']
        nreal_cap = wind_meta_df[wind_meta_df['Facility.Name']==nrel_name]['Capacity'].values[0]

        filename = nrel_name.rstrip().replace(' ', '_').replace('/', '_') + '.csv'
        print(filename)
        
        df = pd.read_csv(Path(datadir, date.strftime('%Y%m%d'), 'wind', filename))
        df.iloc[:, 2:] *= ny_cap / nreal_cap
                
        filename = ny_name.rstrip().replace(' ', '_') + '.csv'
        df.to_csv(Path(savedir, date.strftime('%Y%m%d'), 'wind', filename), index=False)
```

# Solar

```python
solar_mapping = pd.read_csv('~/Downloads/Solar_Mapping_NY_NREL2030.csv')
solar_meta_df = pd.read_csv('~/Research/PERFORM/Data/ForArnab/MetaData/solar_meta.csv')
```

```python
datadir = '~/Research/PERFORM/Data/ForArnab/'
savedir = '~/Research/PERFORM/Data/ForArnab/NYISO/'

for date in pd.date_range(start='2019-08-02', periods=7, freq='D'):
    
    for _, row in solar_mapping.iterrows():
        ny_name = row['name_NY']
        
        nrel_name = row['name_NREL']
        
        ny_cap = row['capacity_NY']
        nreal_cap = solar_meta_df[solar_meta_df['site_ids']==nrel_name]['AC_capacity_MW'].values[0]

        filename = nrel_name.rstrip().replace(' ', '_').replace('/', '_') + '.csv'
        print(filename)
        
        df = pd.read_csv(Path(datadir, date.strftime('%Y%m%d'), 'solar', filename))
        df.iloc[:, 2:] *= ny_cap / nreal_cap
                
        filename = ny_name.rstrip().replace(' ', '_') + '.csv'
        df.to_csv(Path(savedir, date.strftime('%Y%m%d'), 'solar', filename), index=False)
```
